### جلسة 1.1: البتات الكلاسيكية مقابل البتات الكمومية
</div>
<div dir="rtl" style="text-align: right;">
قبل أن نُقدم بإيجاز الحوسبة الكمومية، دعونا نُراجع أولًا الوحدة الأساسية للحوسبة الكلاسيكية — البتات الكلاسيكية.
</div>

<div dir="rtl" style="text-align: right;">
يوجد البت الكلاسيكي في حالتين فقط: "0" أو "1"، ويخضع هذا الأسلوب لمبادئ الفيزياء الكلاسيكية.
</div>
<div dir="rtl" style="text-align: right;">
في المقابل، الوحدة الأساسية في الحوسبة الكمومية هي البت الكّمومي (الكيوبت)، الذي يخزن المعلومات الكمومية. 
</div>
<div dir="rtl" style="text-align: right;">
على عكس البتات الكلاسيكية، يُمكن أن يكون الكيوبت في الحالة $|0\rangle$، أو الحالة $|1\rangle$، أو في حالة تراكب لكلتا الحالتين في آنٍ واحد.
</div>
<div dir="rtl" style="text-align: right;">
تُمثل هذه الحالات عادةً باستخدام ترميز ديراك (كيت) - Dirac (ket) notation، مثل $|0\rangle$ و $|1\rangle$، وهي تتبع مبادئ ميكانيكا الكم
</div>
<div dir="rtl" style="text-align: right;">
ميكانيكا الكم هي " دراسة الفيزياء على نطاقات صغيرة للغاية، وتكشف عن مبادئ الطبيعة الأساسية المدهشة." (Translated from - تم ترجمتها من [1])
</div>
<div dir="rtl" style="text-align: right;">
كما تتمتع البتات الكمومية بخصائص فريدة وقوية أخرى، والتي سنستكشفها في الجلسات القادمة.
</div>

<div dir="rtl" style="text-align: right;">

### جلسة 1.2: التراكب — Superposition
</div>
<div dir="rtl" style="text-align: right;">
 يمكن للبت الكمومي وضع المعلومات الكمومية التي يحملها في تراكب كمومي، والذي يمثل جميع التركيبات الممكنة لتكوينات البت الكمومي. (adapted and translated from - تم ترجمتها واقتباسها من [1])
</div>
<div dir="rtl" style="text-align: right;">
تُعد هذه إحدى المزايا الرئيسية مقارنةً بالبتات الكلاسيكية.
</div>
<div dir="rtl" style="text-align: right;">
تشبيه بسيط يجعلك تفكر بشكل صحيح في التراكب:
</div>
<div dir="rtl" style="text-align: right;">
افترض أن لديك عملة معدنية وقمت بترميزها على النحو التالي - صورة (الحالة $|0\rangle$)، كتابة (الحالة $|1\rangle$) - ثم رميتها في السماء.
</div>
<div dir="rtl" style="text-align: right;">
يحدث التراكب عندما تكون العملة في الهواء: يمكن أن تكون إما صورة ($|0\rangle$)، أو كتابة ($|1\rangle$)، أو كليهما - وهو التراكب - حتى نقيسه (هنا: عندما تسقط العملة في يدك)، وتحصل على إما صورة ($|0\rangle$) أو كتابة ($|1\rangle$).
</div>
<div dir="rtl" style="text-align: right;">
وبالتالي، تنهار حالة التراكب إلى حالة واحدة عند القياس. (adapted and translated from - تم اقتباسها وترجمتها من [1])
</div>
<div dir="rtl" style="text-align: right;">
يتم التعبير عنا رياضيًا على النحو التالي:
</div>
<div dir="rtl" style="text-align: right;">
.$|\psi\rangle = \alpha|0\rangle + \beta|1\rangle$
</div>
<div dir="rtl" style="text-align: right;">
حيثُ أن:
</div>
<div dir="rtl" style="text-align: right;">
$|\psi\rangle$: الدالة الموجية.
</div>
<div dir="rtl" style="text-align: right;">
$\alpha$: المعامل (السعة الاحتمالية) للحالة $|0\rangle$.
</div>
<div dir="rtl" style="text-align: right;">
$\beta$: المعامل (السعة الاحتمالية) للحالة $|1\rangle$.
</div>
<div dir="rtl" style="text-align: right;">
<div dir="rtl" style="text-align: right;">
لن نخوض في تفاصيل الرياضيات هنا، فهذا الدفتر لا يركز على الجانب الرياضي بعمق، بل على فهم المفاهيم الأساسية.
</div>

<div dir="rtl" style="text-align: right;">

### جلسة 1.3: التشابك — Entanglement
</div>
<div dir="rtl" style="text-align: right;">
يُعد التشابك من أهم خصائص الكيوبتات.
</div>
<div dir="rtl" style="text-align: right;">
عندما يكون كيوبتان متشابكين، فإنهما سيكونان في الحالة نفسها، مما يُكوّن لنا نظامًا متشابكًا. (adapted and translated from - تم اقتباسها وترجمتها من [1])
</div>
<div dir="rtl" style="text-align: right;">
عندما نقيس أحد أجزاء النظام المتشابك (الكيوبت المتشابك) لتحديد حالته، يمكننا معرفة حالة الجزء الآخر (الكيوبت المتشابك الآخر) من خلال الكيوبت الذي قمنا بقياسه. (adapted and translated from - تم اقتباسها وترجمتها من [1])
</div>

##### ملاحظة: بعض أمثلة الأكواد مقتبسة ومترجمة من [2] لعرض عدة بوابات كمومية باستخدام دارة كمومية واحدة.

##### Note: Some of the code examples are adapted and translated from [2] to demonstrate multiple quantum gates using a single quantum circuit.

```python
#Importing Required libraries
#استيراد المكتبات المطلوبة
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit import transpile
from qiskit.visualization import plot_histogram


#We give the name qu_circuit(shortcut for quantum circuit) for our quantum circuit.
#Important note:  It doesn't matter the name of the circuit this is only for make it clear for you.

#نسمي الدارة الكمومية الخاصة بنا qu_circuit (اختصار لـ quantum circuit).
#ملاحظة مهمة: لا يهم اسم الدارة، فهذه التسمية فقط لتوضيح الأمور لك.

#Here creating a two qubit quantum circuit
#هنا نقوم بإنشاء دارة كمومية مكونة من كيوبتين (بتين كموميين)
qu_circuit = QuantumCircuit(2)

#رسم الدارة
qu_circuit.draw('mpl') 
```

<div dir="rtl" style="text-align: right;">
    
### جلسة 2.1: بوابة $X$
</div>
<div dir="rtl" style="text-align: right;">
بوابتنا الأولى للمناقشة في هذا الدفتر.
</div>
<div dir="rtl" style="text-align: right;">
بوابة **$X$ (NOT)** هي بوابة كيوبت واحد، وعندما يتم تطبيقها على كيوبت فإنها تقلّب حالته.
</div>
<div dir="rtl" style="text-align: right;">
عند تطبيق بوابة $X$ على الحالة $|0\rangle$، تكون النتيجة: $|1\rangle$. (adapted and translated from - تم اقتباسها وترجمتها من [2])
</div>
<div dir="rtl" style="text-align: right;">
عند تطبيق بوابة $X$ على الحالة $|1\rangle$، تكون النتيجة: $|0\rangle$. (adapted and translated from - تم اقتباسها وترجمتها من [2])
</div>
<div dir="rtl" style="text-align: right;">
إذاً، العملية مشابهة لبوابة الـ NOT الكلاسيكية، والمعروفة أيضًا باسم قَلّب البت. (adapted and translated from - تم اقتباسها وترجمتها من [2])
</div>

```python
#Here is the Python code of applying X gate
#كود (رمز برمجي) بايثون لتطبيق بوابة X.

#First Create our single qubit Quantum Circuit
#أولًا، نقوم بإنشاء دائرة كمومية مكونة من كيوبت واحد
qu_circuit = QuantumCircuit(1)

#Apply X gate on qubit one(indexed as 0) → flip the state from |0⟩ to |1⟩
#تطبيق بوابة X على الكيوبت الأول (معرّف برقم 0) ← قلب الحالة من  ∣0⟩ إلى ∣1⟩ 
qu_circuit.x(0) 

#رسم الدارة
qu_circuit.draw('mpl') 
```

<div dir="rtl" style="text-align: right;">
"في IBM Quantum، الحالة الابتدائية تكون مهيأة على الحالة $|0\rangle$". (translated from - تم ترجمتها من [2])
</div>
<div dir="rtl" style="text-align: right;">

##### ماذا حدث في الأعلى؟
</div>
<div dir="rtl" style="text-align: right;">
ما حدث أعلاه أن الكيوبت مهيأ افتراضيًا في الحالة $|0\rangle$، وقمنا بتطبيق بوابة $X$ عليه، فتم قلب حالته إلى الحالة $|1\rangle$.
</div>

<div dir="rtl" style="text-align: right;">
    
### جلسة 2.2: بوابة هادامارد — Hadamard
</div>
<div dir="rtl" style="text-align: right;">
بوابة هادامارد (H) هي بوابة كيوبت واحد، وعند تطبيقها على كيوبت، فإنها تحوله إلى حالة تراكب كمومي.
</div>
<div dir="rtl" style="text-align: right;">
عند تطبيق بوابة H على الحالة $|0\rangle$، فإنها تُنشئ حالة تراكب بين $|0\rangle$ و $|1\rangle$، ويُعبر عن ذلك رياضيًا كما يلي: $H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}} \quad$ (adapted and translated from - تم اقتباسها وترجمتها من [2])
</div>
<div dir="rtl" style="text-align: right;">
عند تطبيق بوابة H على الحالة $|1\rangle$، فإنها تُنشئ حالة تراكب بين $|0\rangle$ و $|1\rangle$، ويُعبر عن ذلك رياضيًا كما يلي: $H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}$ (adapted and translated from - تم اقتباسها وترجمتها من [2])
</div>

```python
#Here is the python code of applying H gate
#كود (رمز برمجي) بايثون لتطبيق بوابة H.

#First Create our single qubit Quantum Circuit
#أولًا، نقوم بإنشاء دائرة كمومية مكونة من كيوبت واحد
qu_circuit = QuantumCircuit(1)

#Apply Hadamard gate on qubit one(indexed here as (0)) → creates superposition
# تطبيق بوابة هادامارد على الكيوبت الأول (معرّف برقم 0) ← إنشاء حالة تراكب كمومي
qu_circuit.h(0)

#رسم الدارة
qu_circuit.draw('mpl')
```

<div dir="rtl" style="text-align: right;">

</div>
<div dir="rtl" style="text-align: right;">

</div>
<div dir="rtl" style="text-align: right;">
هذا يؤدي إلى إنشاء حالة تراكب تجمع بين $|0\rangle$ و $|1\rangle$ في الوقت نفسه.
</div>
<div dir="rtl" style="text-align: right;">
يُعبر عن ذلك رياضيًا (من [2]) كما يلي:  $H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}} \quad$
</div>

### جلسة 2.3: بوابة CNOT

"بوابة CNOT (وتُعرف أيضًا باسم Controlled NOT أو CX) هي بوابة مكونة من كيوبتين، أي أن عملها يتضمن كيوبتين في وقت واحد: كيوبت التحكم (controlled qubit)  والكيوبت المستهدف (target qubit) ". (translated from - تم ترجمتها من [2])

<div dir="rtl" style="text-align: right;">
تقوم بوابة CNOT بقلّب (تطبيق بوابة X (NOT) على) كيوبت الهدف فقط إذا كانت حالة كيوبت التحكم هي $|1\rangle$. (adapted and translated from - تم اقتباسها وترجمتها من [2])

```python
#Here is the Python code of applying CNOT gate
#كود (رمز برمجي) بايثون لتطبيق بوابة CNOT.

#First Create our two qubit Quantum Circuit
#هنا نقوم بإنشاء دارة كمومية مكونة من كيوبتين (بتين كموميين)
qu_circuit = QuantumCircuit(2)

#Apply CNOT gate → (Control qubit, target qubit)
# تطبيق بوابة CNOT → (الكيوبت المستهدف ,كيوبت التحكم)
qu_circuit.cx(0,1) 

#رسم الدارة
qu_circuit.draw('mpl') 
```

<div dir="rtl" style="text-align: right;">
عند قياس الكيوبتين مرات عديدة (shots)، نتوقع الحصول على إما $|00\rangle$ أو $|11\rangle$ باحتمالية متساوية تقريبًا.
</div>
<div dir="rtl" style="text-align: right;">
هذا يعني:
</div>
<div dir="rtl" style="text-align: right;">
إذا تم قياس الكيوبت الأول على الحالة $|0\rangle$، فسيكون الكيوبت الثاني أيضًا في الحالة $|0\rangle$.
</div>
<div dir="rtl" style="text-align: right;">
إذا تم قياس الكيوبت الأول على الحالة $|1\rangle$، فسيكون الكيوبت الثاني أيضًا في الحالة $|1\rangle$.
</div>
<div dir="rtl" style="text-align: right;">
هذه النتائج المرتبطة تمثل خاصية التشابك الكمومي، وهي السمة المميزة له.
</div>
